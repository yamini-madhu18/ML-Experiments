```python
import pandas as pd
import matplotlib.pyplot as plt

print("Name: Yamini M")
print("Roll No: 24BAD131")

df = pd.read_csv("data.csv", encoding="ISO-8859-1")

print(df.head())
print(df.tail())
print(df.info())
print(df.describe())

print(df.isnull().sum())

sales_per_product = df.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)

sales_per_product.plot(kind="bar")
plt.title("Top 10 Products by Sales Quantity")
plt.show()

sales_trend = df.groupby("InvoiceDate")["Quantity"].sum().head(50)
sales_trend.plot(kind="line")
plt.title("Sales Trend")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Name: Yamini M")
print("Roll No: 24BAD131")

df = pd.read_csv("diabetes.csv")

print(df.head())
print(df.info())
print(df.isnull().sum())

plt.hist(df["Glucose"], bins=20)
plt.title("Glucose Level Distribution")
plt.show()

plt.boxplot(df["Age"])
plt.title("Age Distribution")
plt.show()

sns.pairplot(df)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Name: Yamini M")
print("Roll No: 24BAD131")

df = pd.read_csv("housing.csv")

print(df.head())
print(df.info())
print(df.isnull().sum())

sns.scatterplot(x="area", y="price", data=df)
plt.show()

corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Name: Yamini M")
print("Roll No: 24BAD131")

df = pd.read_csv("marketing_campaign.csv", sep="\t")

print(df.head())
print(df.info())
print(df.isnull().sum())

df["Age"] = 2024 - df["Year_Birth"]

plt.bar(df["Education"].value_counts().index, df["Education"].value_counts().values)
plt.title("Education Distribution")
plt.xticks(rotation=45)
plt.show()

plt.boxplot(df["Income"].dropna())
plt.title("Income Distribution")
plt.show()

plt.boxplot(df["MntWines"])
plt.title("Wine Spending Pattern")
plt.show()
```
